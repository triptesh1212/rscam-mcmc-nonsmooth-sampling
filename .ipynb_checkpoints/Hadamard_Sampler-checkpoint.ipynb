```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf
```

## Hadamard for the 1D Bayesian LASSO Posterior

We consider the Lasso posterior $\pi(x) \propto \exp\Big(-\beta (\tfrac{1}{2} \|Ax - y\|_2^2 + \lambda \|x\|_1) \Big)$

We introduce latent variables $\eta > 0$ and obtain the density

$$ 
\tilde{\pi}(x,\eta)
\propto
\exp\!\left(
-
\left(
\frac{x^2}{2\eta}
+
\frac{\beta^2 \lambda^2 \eta}{2}
\right)
-
\frac{\beta}{2}\|Ax-y\|^2
\right)
$$

By applying a change of variables of $x = u \odot v$ and $\eta = u \odot u/(\lambda \beta)$, we get the following desnsity

$$
\pi(u, v) \propto u \exp\left(-\beta\left(\frac{1}{2}\lambda\ (u^2 + v^2) + \frac{1}{2}(Auv - y)^2\right)\right)
$$

The gradient of $G(x)$ with respect to $x$ is $\nabla G(x) = 2A(Ax - y)$. Substituting $x = uv$ gives $\nabla G(uv) = 2A(Auv - y)$

The time-stepping equations are

$$
u^{k+1} - u^k
=
-\lambda u^{k+1} \Delta t
-
2A v^k (A u^k v^k - y)\Delta t
+
\frac{\Delta t}{\beta u^{k+1}}
+
\sqrt{\frac{2}{\beta}}\,\Delta W_k^{1}
$$

$$
v^{k+1} - v^k
=
-\lambda v^{k+1} \Delta t
-
2A u^k (A u^k v^k - y)\Delta t
+
\sqrt{\frac{2}{\beta}}\,\Delta W_k^{2}.
$$


Rearranging gives

$$
v^{k+1}
=
\frac{
v^k
-
2A u^k (A u^k v^k - y)\Delta t
+
\sqrt{\frac{2}{\beta}}\,\Delta W_k^{2}
}{
1 + \lambda \Delta t
}
$$

$$
(1+\lambda\Delta t)(u^{k+1})^2
-
C u^{k+1}
-
\frac{\Delta t}{\beta}
=0
$$

where, $C
=
u^k
-
2A v^k (A u^k v^k - y)\Delta t
+
\sqrt{\frac{2}{\beta}}\,\Delta W_k^{1}$


Taking the positive root,
$$
u^{k+1}
=
\frac{
C + \sqrt{C^2 + \frac{4(1+\lambda\Delta t)\Delta t}{\beta}}
}{
2(1+\lambda\Delta t)
}
$$

```python
def hadamard(Beta, Lambda, step, n_samples, burn_in):

    np.random.seed(0)

    A = 1.0
    y = 1.0
    dt = step

    samples = []

    u = np.zeros(n_samples)
    v = np.zeros(n_samples)

    # initial values
    u[0] = 1.0
    v[0] = 1.0

    for k in range(n_samples-1):
        
        dW1 = np.sqrt(dt) * np.random.randn()
        dW2 = np.sqrt(dt) * np.random.randn()
    
        gradG = 2*A*(A*u[k]*v[k] - y)
    
        num = v[k] - 2*A*u[k]*(A*u[k]*v[k] - y)*dt + np.sqrt(2/Beta)*dW2
        v[k+1] = num/(1 + Lambda*dt)
    
        C = u[k] - 2*A*v[k]*(A*u[k]*v[k] - y)*dt + np.sqrt(2/Beta)*dW1
    
        a = (1 + Lambda*dt)
        b = -C
        c = -dt/Beta
    
        disc = b**2 - 4*a*c
        u_new = (-b + np.sqrt(disc))/(2*a)
    
        u[k+1] = u_new
        
        samples.append(u[k+1] * v[k+1])


    return samples
```

```python
def exact_posterior(x):
    x = np.asarray(x)

    Z = (np.sqrt(np.pi) / 2) * (np.exp(-1) + np.exp(3) * (1 - erf(2)))

    pdf = np.zeros_like(x, dtype=float)

    pos = x >= 0
    neg = x < 0

    pdf[pos] = np.exp(-1) * np.exp(-x[pos]**2)
    pdf[neg] = np.exp(3) * np.exp(-(x[neg] - 2)**2)

    return pdf / Z
```

```python
samples_hadamard = hadamard(Beta=2, Lambda=1, step=0.5, n_samples=50000, burn_in=5000)

x_plot = np.linspace(-2, 3, 10000)
y_exact = exact_posterior(x_plot)

plt.figure(figsize=(10, 6))
plt.hist(samples_hadamard, bins=100, density=True, alpha=0.7, color='steelblue', edgecolor='black')
plt.plot(x_plot, y_exact, 'r-', linewidth=2.5, label='Exact Posterior')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Posterior Distribution using Hadamard sampler')
plt.grid(True, alpha=0.3)
plt.show()
```
